# Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
df = pd.DataFrame(np.random.normal(0,10,size=(10000, 2)), columns=list('AB'))
df.loc[:,'C'] = list(np.random.normal(-50,10,size=(5000,1))) + list(np.random.normal(10,10,size=(5000,1)))
```

```python
df.loc[:,'A'] = df.A.astype(int)
df.loc[:,'B'] = df.B.astype(int)
df.loc[:,'C'] = df.C.astype(int)
```

```python
sns.heatmap(data=df.loc[:,['A','C']].pivot_table(index='A', columns='C', aggfunc=len))
```
